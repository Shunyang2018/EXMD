```python
import pandas as pd
import numpy as np
import os

import spectral_entropy
```

```python
def msp2list(msp):
    with open(msp) as f:
        for block in readblocks(f):
            x=[]
            for line in block:
                line = line.strip('\n')
                line = line.split(';')
                for pair in line:
                    pair = pair.split()
                    if len(pair) == 2:
                        x.append([float(pair[0]),float(pair[1])])
    return x

def jdx2list(filename):
    '''
    read and clean computational mass spectrum result jdx file
    Input
        filename: str, name and path of the jdx file
    Output
        x: normalized mass spectrum
    '''
    with open(filename)as f:
        tmp = f.readlines()
    x = []
    for line in tmp[7:-1]:
        pair = line.split()
        x.append([float(pair[0]),float(pair[1])])
    return x

def index2list(massindex, Index):    
    spec=Index[Index['Database No.'] == massindex]['MASS']
    spec = spec.iloc[0].split()[1:-1]
    x = []
    for i in range(len(spec)):
        if float(spec[i]) != 0:
            x.append([i,spec[i]])
    return x
```

```python
path = './om2/'
indexfile = '../QCEIMS_SI.xlsx'
Index = pd.read_excel(indexfile,sheet_name='data',header=[0],index_col=None)
```

```python

files= os.listdir(path)
smi_list = []
index = []
for file in files:
    print(file)
    if os.path.splitext(file)[1] == ".jdx":
        query = jdx2list(path+file)
        massindex = int(file.split('/')[-1].split('.')[0])
        try:
            ref = index2list(massindex, Index)
            smi = spectral_entropy.similarity(query, ref,method='entropy',ms2_da=0.05)
            smi_list.append(smi)
            
        except:
            smi_list.append('NA')
        index.append(massindex)
            

df = pd.DataFrame({'index':index,'entropy':smi_list})
df.to_csv('om2_entropy.csv')
```
